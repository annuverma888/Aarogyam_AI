```python
import pandas as pd

description_df = pd.read_csv("../dataset/disease_description.csv")

print(description_df.head())
print(description_df.shape)
```

```python
def get_description(disease):

    result = description_df[
        description_df["Disease"].str.lower() == disease.lower()
    ]

    if result.empty:
        return "Description not available."

    return result.iloc[0]["Description"]
```

```python
print(get_description("Diabetes"))
```

```python
print(get_description("Malaria"))
```

```python
disease = input("Enter Disease Name: ")

print("\nDisease:", disease)
print("\nDescription:")
print(get_description(disease))
```

```python
import joblib
import numpy as np

model = joblib.load("../model/disease_model.pkl")
encoder = joblib.load("../model/label_encoder.pkl")
```

```python
training_df = pd.read_csv("../dataset/Training.csv")

# Remove unwanted column if present
if "Unnamed: 133" in training_df.columns:
    training_df.drop(columns=["Unnamed: 133"], inplace=True)

X = training_df.drop("prognosis", axis=1)

sample = X.iloc[0].values.reshape(1, -1)

prediction = model.predict(sample)

predicted_disease = encoder.inverse_transform(prediction)[0]

print("Predicted Disease:", predicted_disease)

print("\nDescription:")
print(get_description(predicted_disease))
```
