```python
def get_medicines(disease):

    disease = disease.strip().lower()

    row = medicine_df[
        medicine_df["Disease"].str.lower() == disease
    ]

    if row.empty:
        return []

    medicines = []

    for col in ["Medicine_1", "Medicine_2", "Medicine_3"]:

        med = row.iloc[0][col]

        if pd.notna(med):
            medicines.append(med)

    return medicines
```

```python
print(get_medicines("Diabetes"))
```

```python
model = joblib.load("../model/disease_model.pkl")

encoder = joblib.load("../model/label_encoder.pkl")
```

```python
train = pd.read_csv("../dataset/Training.csv")

if "Unnamed: 133" in train.columns:
    train.drop(columns=["Unnamed: 133"], inplace=True)

X = train.drop("prognosis", axis=1)
```

```python
sample = X.iloc[25].values.reshape(1, -1)

prediction = model.predict(sample)

disease = encoder.inverse_transform(prediction)[0]

print("Predicted Disease :", disease)
```

```python
medicines = get_medicines(disease)

print("\nRecommended Medicines\n")

for i, medicine in enumerate(medicines, 1):
    print(f"{i}. {medicine}")
```
