```python
import pandas as pd
import joblib
```

```python
precaution_df = pd.read_csv("../dataset/symptom_precaution.csv")

precaution_df.head()
```

```python
print("Shape:", precaution_df.shape)
print("\nColumns:")
print(precaution_df.columns)
```

```python
def get_precautions(disease):

    disease = disease.strip().lower()

    row = precaution_df[
        precaution_df["Disease"].str.lower() == disease
    ]

    if row.empty:
        return []

    precautions = row.iloc[0,1:].dropna().tolist()

    return precautions
```

```python
print(get_precautions("Diabetes"))
```

```python
model = joblib.load("../model/disease_model.pkl")

encoder = joblib.load("../model/label_encoder.pkl")
```

```python
train = pd.read_csv("../dataset/Training.csv")

if "Unnamed: 133" in train.columns:
    train.drop(columns=["Unnamed: 133"], inplace=True)

X = train.drop("prognosis", axis=1)
```

```python
sample = X.iloc[100].values.reshape(1,-1)

prediction = model.predict(sample)

disease = encoder.inverse_transform(prediction)[0]

print("Predicted Disease:", disease)
```

```python
precautions = get_precautions(disease)

print("\nRecommended Precautions:\n")

for i, precaution in enumerate(precautions, start=1):
    print(f"{i}. {precaution}")
```
