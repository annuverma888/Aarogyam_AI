```python
import pandas as pd
import numpy as np
import joblib
```

```python
model = joblib.load("../model/disease_model.pkl")
encoder = joblib.load("../model/label_encoder.pkl")
```

```python
description_df = pd.read_csv("../dataset/disease_description.csv")
precaution_df = pd.read_csv("../dataset/symptom_precaution.csv")
medicine_df = pd.read_csv("../dataset/medicine_dataset.csv")
diet_df = pd.read_csv("../dataset/diet_dataset.csv")
workout_df = pd.read_csv("../dataset/workout_dataset.csv")

train = pd.read_csv("../dataset/Training.csv")

if "Unnamed: 133" in train.columns:
    train.drop(columns=["Unnamed: 133"], inplace=True)

X = train.drop("prognosis", axis=1)
```

```python
def get_description(disease):

    row = description_df[
        description_df["Disease"].str.lower()==disease.lower()
    ]

    if row.empty:
        return "Not Available"

    return row.iloc[0]["Description"]
```

```python
def get_precautions(disease):

    row = precaution_df[
        precaution_df["Disease"].str.lower()==disease.lower()
    ]

    if row.empty:
        return []

    return row.iloc[0,1:].dropna().tolist()
```

```python
def get_medicines(disease):

    row = medicine_df[
        medicine_df["Disease"].str.lower()==disease.lower()
    ]

    if row.empty:
        return []

    return row.iloc[0,1:].dropna().tolist()
```

```python
def get_diet(disease):

    row = diet_df[
        diet_df["Disease"].str.lower()==disease.lower()
    ]

    if row.empty:
        return {}

    return {
        "Breakfast":row.iloc[0]["Breakfast"],
        "Lunch":row.iloc[0]["Lunch"],
        "Dinner":row.iloc[0]["Dinner"],
        "Avoid":row.iloc[0]["Avoid"]
    }
```

```python
def get_workout(disease):

    row = workout_df[
        workout_df["Disease"].str.lower()==disease.lower()
    ]

    if row.empty:
        return {}

    return {
        "Exercise 1":row.iloc[0]["Exercise_1"],
        "Exercise 2":row.iloc[0]["Exercise_2"],
        "Exercise 3":row.iloc[0]["Exercise_3"],
        "Duration":row.iloc[0]["Duration"]
    }
```

```python
def predict_disease(selected_symptoms):

    input_data = np.zeros(len(X.columns))

    for symptom in selected_symptoms:

        if symptom in X.columns:
            index = X.columns.get_loc(symptom)
            input_data[index] = 1

    prediction = model.predict([input_data])

    disease = encoder.inverse_transform(prediction)[0]

    return disease
```

```python
symptoms = [
    "itching",
    "skin_rash",
    "nodal_skin_eruptions"
]

disease = predict_disease(symptoms)

print("Disease:",disease)
```

```python
print("="*60)

print("Predicted Disease")
print(disease)

print("="*60)

print("Description")
print(get_description(disease))

print("="*60)

print("Medicines")

for med in get_medicines(disease):
    print("•",med)

print("="*60)

print("Precautions")

for p in get_precautions(disease):
    print("•",p)

print("="*60)

diet=get_diet(disease)

print("Diet")

for k,v in diet.items():
    print(k,":",v)

print("="*60)

work=get_workout(disease)

print("Workout")

for k,v in work.items():
    print(k,":",v)
```
